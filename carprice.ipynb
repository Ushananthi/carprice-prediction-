```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("CarPrice_Assignment.csv")
```

```python
df.info()
```

```python
df.corr()
```

```python
df.drop(["drivewheel","stroke","doornumber","carbody","car_ID","wheelbase","cylindernumber","compressionratio","fueltype","boreratio","symboling","fuelsystem"],axis=1,inplace=True)
```

```python
df.drop(["CarName","aspiration","enginelocation","carheight"],axis=1,inplace=True)
```

```python
df
```

```python
sns.boxplot(data=df);
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordi=OrdinalEncoder()
df["enginetype"]=ordi.fit_transform(df[["enginetype"]])
```

```python
df.corr()
```

```python
df.drop(["enginetype","peakrpm"],axis=1,inplace=True)
```

```python
q1=np.quantile(df["price"],0.25)
q3=np.quantile(df["price"],0.75)
iqr=q3-q1
uw=q3+1.5*iqr
lw=q1-1.5*iqr
for i in df["price"]:
    if i>uw:
        df["price"]=df["price"].replace(i,uw)
```

```python
sns.boxplot(data=df)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
print(rmse)
```

```python
lr.score(xtrain,ytrain)
```
